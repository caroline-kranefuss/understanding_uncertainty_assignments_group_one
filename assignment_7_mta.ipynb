1.1. The expected value of a single die roll is 3.5. This is based on taking the average of all possible outcomes: 

$$
\frac{1 + 2 + 3 + 4 + 5 + 6}{6} = 3.5.
$$

1.2. The expected value of the sum of two dice rolls is 7, assuming independence. This is because:

$$
E(X + Y) = E(X) + E(Y) = 3.5 + 3.5 = 7.
$$

1.3 In European roulette, there are 36 slots for the gambler to bet on, but **37 possible outcomes** (the extra slot being the green 0).  
This creates an inherent disadvantage to the gambler. Normally, the expected value of a bet would be $ \frac{1}{36} $ scaled by the wager,  
but in European roulette it becomes $ \frac{1}{37} $.

For a \$1 bet on a single color (red or black), the expected value per bet is:

$$
E[X] = \frac{18}{37}(+1) + \frac{19}{37}(-1) = \frac{18 - 19}{37} = -\frac{1}{37}.
$$

Thus, for every \$1 bet, the player loses on average:

$$
\boxed{E[X] = -0.027} \quad \text{(equivalently, a 2.7\% (-\$0.027) expected loss per spin).}
$$
